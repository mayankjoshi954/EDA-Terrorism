## 2. Cleaning the Data

```python
#dropping those columns which have null values greater than 10000
for x in df.columns:
    if df[x].isna().sum()>10000:
        df.drop(x,axis =1,inplace=True)
```

```python
df.isna().sum().sort_values(ascending = False)
```

```python
df.head()
```

```python
#now check the availabe columns we have
df.columns.values
```

```python
#seelcting columns to remove
to_remove = ['eventid','imonth','iday','extended','country','region','vicinity', 'crit1',
       'crit2', 'crit3', 'doubtterr', 'multiple','specificity','suicide',
       'attacktype1','targtype1','target1', 'natlty1', 'natlty1_txt', 'gname', 'guncertain1',
       'individual', 'weaptype1','property',
       'ishostkid', 'dbsource', 'INT_LOG', 'INT_IDEO', 'INT_MISC',
       'INT_ANY']
```

```python
#removing columns which are not required
for i in df.columns:
    if(i in to_remove):
        df.drop(i,axis =1,inplace=True)
```

```python
df.head()
```

```python
#changing columns names to more readable form
df.columns = ['Year','Country','Region','Province/State','City','Latitude','Longitude','Success','Attack Type','Target Type','Weapon Type']
```

```python
df.isna().sum()
```

```python
#filling up null values
df['City'] = df['City'].fillna('Unknown')
df['Province/State'] = df['Province/State'].fillna('Unknown')
```

```python
df.isna().sum()
```
